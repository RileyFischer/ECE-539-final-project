```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("breast-cancer.csv")
#encode our output variable 
df['diagnosis'] = df['diagnosis'].map({'B' : 0, 'M' : 1})
df['diagnosis'] = pd.to_numeric(df['diagnosis'])
df
```

```python
corr_mat = df.corr(method='spearman')
corr_mat
```

```python
Groups={}
final_features=[]
for i in corr_mat:
    Groups[i]=list(corr_mat[(abs(corr_mat[i]) >= 0.9) & (corr_mat.columns != i)].index)
    if Groups[i]==[]:
        Groups.pop(i)
        final_features.append(i)

sorted_groups={}
for i in sorted(Groups, key=lambda k: len(Groups[k]), reverse=True):
    sorted_groups[i]=Groups[i]

while sorted_groups!={}:
    print("pick desired features from list, preferably ones in the begining of the list: ",sorted_groups.keys())
    remove=input()
    
    try:
        for i in sorted_groups.pop(remove):
            try:
                sorted_groups.pop(i)
            except:
                pass
    except:
        print("try again")
    final_features.append(remove)
final_features
```
